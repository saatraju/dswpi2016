```python
%matplotlib inline
import json
import pandas as pd
from IPython.display import display
from collections import Counter
import matplotlib.pyplot as plt

tweets_data_path = 'trump_tweets.json'
pd.set_option('display.mpl_style', 'default') #Pretty print graph

tweets_data = []
RT_tweets = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)        
        tweets_data.append(tweet)
    except:
        continue
print "Total Tweet Count : "+str(len(tweets_data) )

#Function to convert JSON data to panda Datasets
def jsonto_PandaDataFrame(tweets):
    import pandas as pd
    #Create a instance of Panda Dataframe
    DataSet = pd.DataFrame()    
    #print tweets
    for item in tweets:
        DataSet['word'] = [l[0] for l in tweets]
        DataSet['Count'] = [l[1] for l in tweets]
    return DataSet

def print_frequency_count(item_list):    
    c = Counter(item_list)
    result = c.most_common()[:1000]
    DataSet = jsonto_PandaDataFrame(result)
    display(DataSet.head(50))
    
def get_word_frequency():
   
    status_texts = [ status['text'] 
                    for status in tweets_data ]

    screen_names = [ user_mention['screen_name']
                    for status in tweets_data
                        for user_mention in status['entities']['user_mentions'] ]

    hashtags = [ hashtag['text'] 
             for status in tweets_data
                 for hashtag in status['entities']['hashtags'] ]
    # Compute a collection of all words from all tweets
    words = [ w 
             for t in status_texts 
              for w in t.split() if not w.startswith(('#', '@')) ]
    
    print "***** PRINTING TOP 50 WORD COUNT *****"
    print_frequency_count(words)
    print "***** PRINTING TOP 50 SCREEN NAMES COUNT *****"
    print_frequency_count(screen_names)
    print "***** PRINTING TOP 50 HASHTAG COUNT *****"
    print_frequency_count(hashtags)      
        

def plot_graph(tzs):
    # Create a bar-graph figure of the specified size
    plt.rcParams['figure.figsize'] = (15, 5)
    # Plot the Time Zone data as a bar-graph
    tzs.plot(kind='bar')
    # Assign labels and title to the graph to make it more presentable
    plt.xlabel('ID')
    plt.ylabel('Retweet_CT')
    plt.title('Top 10 retweets')
    

def get_popular_tweet():
    #Create a instance of Panda Dataframe
    DataSet = pd.DataFrame()    
    DataSet['ID'] = [status["id_str"] for status in tweets_data]
    DataSet['Retweet_CT'] =  [status["retweet_count"] for status in tweets_data]
    DataSet['Text'] =  [status["text"] for status in tweets_data]
    DataSet.sort_values(['Retweet_CT'], ascending=[True],inplace=True)
    #print DataSet
    #display(DataSet)
    tzs = DataSet['Retweet_CT'].value_counts()[:10]
    print tzs
    #plot_graph(tzs)
    retweet_groups = DataSet.groupby('Retweet_CT')
    RT = retweet_groups.size()[:50]
    plot_graph(RT)
#get_popular_tweet()
get_word_frequency()
```
